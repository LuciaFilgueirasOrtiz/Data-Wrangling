***Carga y aproximación a la base de datos***

```python
#importamos las librerías que necesitamos 
import pandas as pd
import numpy as np
```

```python
#cargamos nuestras bases de datos
df1 = pd.read_csv('/content/p_data.csv',delimiter=';')
df2 = pd.read_csv('/content/w_data.csv',delimiter=';')
```

```python
#vemos los primeros registros de la base 
df1.head()
```

```python
#vemos cómo se componen ambos dataset 
print('El df1 se compone de', df1.shape[0], 'filas y de', df1.shape[1], 'columnas')
```

```python
print('El df1 se compone de', df2.shape[0], 'filas y de', df2.shape[1], 'columnas')
```

Vemos que el segundo dataframe tiene un registro más y dos columnas menos. 

***Unimos las dos tablas***

```python
#eliminamos los duplicados de acuerdo a la columna id del df1 
df1.drop_duplicates('ID')
```

```python
#eliminamos los duplicados de acuerdo a la columna id del df2 
df2.drop_duplicates('ID')
```

```python
#realizamos la unión de los dos dataframes con un right merge ya que el segundo df tiene un registro más que el primero 
df_complete = pd.merge(df1, df2, on = 'ID', how='right', indicator=True)
print(f'El dataframe completo tiene {len(df_complete)} filas y {df_complete.shape[1]} columnas')
```

```python
df_complete.head()
```

***Data Wrangling***

Establecemos el ID como índice y lo ordenamos de menor a mayor

```python
df_complete.set_index('ID', inplace=True)
```

```python
df_complete = df_complete.sort_index()
```

```python
df_complete
```

```python
#transformamos los nombres de las columnas a minuscula 
df_complete = df_complete.rename(columns={'Age': 'age', 'Sex': 'sex', 'Race': 'race', 'Country': 'country', 'Education': 'education', 'Target': 'target',
                                          'Education-Num': 'education-Num', 'Marital Status' : 'marital status','Relationship' : 'relationship',
                                          'Workclass': 'workclass','Capital Gain': 'capital gain', 'Capital Loss': 'capital loss', 'Hours per week': 'hours per week', '_merge': 'merge'})
df_complete
```

Para comparar cuales eran los registros que compartían el mismo ID, en el merge creamos una nueva columna que devuelve si el registro están en ambos df o left/right. 

```python
not_in_value = df_complete[df_complete['merge'] =='right_only']
not_in_value

```

Como vemos el ID 1344995623 es el que no tienen en común los dos df. 

Transformamos los target a dos valores únicamente (‘<=50k’ y ‘>50k’) para poder analizarlos posteriormente.

```python
#vemos cuáles son los distintos tipos de target que existen 
df_complete["target"].unique()
```

```python
df_complete['target'].replace([(' >50K.'), (' >50K')], '>50k', inplace=True)
df_complete['target'].replace([(' <=50K'), (' <=50K.')], '<=50k', inplace=True)
```

```python
#comprobamos que los valores que quedaron fueron esos 
df_complete['target'].unique()
```

```python
#vemos cuántos registros quedaron con la separación de target que realizamos 
df_complete.groupby(by='target').count()
```

Transormamos las columnas de sexo, raza y educación para poder analizarla posteriormente. 

```python
#comenzamos viendo cuántos valores distintos existen en la columna race
df_complete['race'].unique()
```

```python
#modificamos los valores para que estén en minúscula y sin espacios en blanco
df_complete = df_complete.replace({'race': {' White':'white',
                                   ' Other':'other',
                                   ' Black':'black',
                                   ' Asian-Pac-Islander': 'asian-pac-islander',
                                   ' Amer-Indian-Eskimo': 'amer-indian-eskimo'}})
```

```python
#comprobamos si los valores se modificaron 
df_complete['race'].unique()
```

```python
#continuamos modificando los valores de sexo, primero vemos cuáles son los valores existentes
df_complete['sex'].unique()
```

```python
#no existe variablilidad en la columna de sexo, no es necesario unificar. Unicamente borraremos los espacios en blanco y tranformalo a minuscula
df_complete = df_complete.replace({'sex': {
                                   ' Male':'male',
                                   ' Female': 'female'}})
```

```python
#comprobamos que los cambios se realizaron correctamente
df_complete['sex'].unique()
```

```python
#vemos los valores que están en la columna educación 
df_complete['education'].unique()
```

```python
#son varios los valores a los que hay quitarle el espacio y pasarlo a miniscula
df_complete['education'] = df_complete['education'].str.lower()
df_complete['education'] = df_complete['education'].str.strip()
```

```python
#comprobamos haber modificado los valores correctamente
df_complete['education'].unique()
```

Estando ya transformadas todas las columnas que necesitamos para el análisis (race, sex, education y target) procedemos a realiazarlo.
